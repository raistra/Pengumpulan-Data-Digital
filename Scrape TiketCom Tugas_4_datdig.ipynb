```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
from urllib.request import urlopen
```

```python
def scrape(urls):
    
    for url in urls:

        request_ = urlopen(url)
        json_fmt = json.loads(request_.read())['data']['departJourneys']
        raw_data = json_fmt['journeys']
        
        for i in raw_data :
            
            information_ = i['segmentSchedules'][0]
            trip_id.append(information_['tripId'])
            trip_date.append(information_['tripDate'])
            train_name.append(information_['trainName'])
            train_number.append(information_['trainNumber'])
            dept_station.append(information_['departureStation']['name'])
            arr_station.append(information_['arrivalStation']['name'])
            dept_date.append(information_['departureDate'])
            arr_date.append(information_['arrivalDate'])
            dept_time.append(information_['departureTime'])
            arr_time.append(information_['arrivalTime'])
            wagon_class.append(information_['wagonClass']['detail'])
            avail_seats.append(information_['availableSeats'])
            pass_type.append(information_['scheduleFares'][0]['paxType'])
            fares.append(information_['scheduleFares'][0]['priceAmount'])
```

```python
if __name__ == "__main__" :
    
    trip_id = []
    trip_date = []
    train_name = []
    train_number = []
    dept_station = []
    arr_station = []
    dept_date = []
    arr_date = []
    dept_time = []
    arr_time = []
    wagon_class = []
    avail_seats = []
    pass_type = []
    fares = []
    urls = ["https://m.tiket.com/ms-gateway/tix-train-search-v2/journeys?acount=1&ddate=20230408&dest=SBI&dtype=STATION&icount=0&orig=GMR&otype=STATION&rdate=&ttype=ONE_WAY",
           "https://m.tiket.com/ms-gateway/tix-train-search-v2/journeys?acount=1&ddate=20230408&dest=YK&dtype=STATION&icount=0&orig=Jakarta&otype=CITY&rdate=&ttype=ONE_WAY"]
    
    scrape(urls)
    
    df = pd.DataFrame({"Trip_ID" : trip_id, "Trip_Date" : trip_date, "Train_Name" : train_name, "Train_Number" : train_number, "Departure_Station" : dept_station, "Arrival_Station" : arr_station,
                   "Departure_Date" : dept_date, "Arrival_Date" : arr_date, "Departure_Time" : dept_time, "Arrival_Time" : arr_time, "Wagon_Class" : wagon_class, "Available_Seats" : avail_seats,
                   "Passenger_Type" : pass_type, "Fares" : fares})
```

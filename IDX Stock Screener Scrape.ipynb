```python
# Mengimport library yg dibutuhkan
import requests
from selenium import webdriver 
import json
import re
import pandas as pd
```

```python
driver = webdriver.Chrome(executable_path= 'chromedriver.exe')
```

```python
# Mengambil API saham harian dari website idx.com

url = 'https://idx.co.id/support/stock-screener/api/v1/stock-screener/get' 

# Melakukan request ke URL
driver.get(url)
```

```python
# Mengambil HTML dari halaman
html_halaman = driver.page_source
html_halaman
```

```python
driver.quit()
```

```python
json_str = html_halaman

# Menghapus tag HTML 
json_str = re.sub(r'<.*?>', '', json_str)

# Mengubah string menjadi JSON
data = json.loads(json_str)

# Mencetak JSON
print(data)
```

```python
df = pd.DataFrame(data['results'])
df
```

```python
# Create new dataframe with desired columns
new_df = df[['companyName', 'stockCode', 'indexCode', 'per', 'pbv']]

# Display new dataframe
print(new_df)
```

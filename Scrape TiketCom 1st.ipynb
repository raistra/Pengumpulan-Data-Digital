## WEBSCRAPING WEBSITE [TIKET.COM](https://www.tiket.com/)
Tahap yang dilakukan:
1.   Mengakses website Tiket.com
2.   Melakukan pencarian tiket kereta api dari Jakarta ke Bandung untuk tanggal 1 April 2023
3.   Setelah ditemukan data, akan dicari link API yang kemudian dibaca oleh python



### Import libraries

```python
import urllib.request
import json
import pandas as pd
```

### Mengakses data melalui link API

```python
url = "https://www.tiket.com/ms-gateway/tix-train-search-v2/journeys?orig=Jakarta&otype=CITY&dest=Bandung&dtype=CITY&ddate=20230401&rdate=&acount=1&icount=0&ttype=ONE_WAY"

with urllib.request.urlopen(url) as response:
    data = json.loads(response.read().decode())

print(data)
```

### Mengubah data dalam format dataframe

```python
df = pd.json_normalize(data, record_path=["data", "departJourneys", "journeys"], meta=[["data", "parameter", "originCode"], ["data", "parameter", "destinationCode"]], errors="ignore")
print(df.head())
```

### Melihat informasi data yang diperoleh

```python
df.info()
```

### Menyimpan data dalam bentuk CSV

```python
df.to_csv('6_Tugas4.csv', index=False)
```
